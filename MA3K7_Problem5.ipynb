# Straightforward Code to quickly compute sums and observe proprties of our construction

```python
# imports
import matplotlib.pyplot as plt
from math import log 
import operator
from itertools import permutations
```

```python
elements=["rightarrow","leftarrow","uparrow","downarrow"]
all_permutations = list(permutations(elements))

for element in all_permutations: 
    ops = {
    f"{element[0]}": operator.add,
    f"{element[1]}": operator.sub,
    f"{element[2]}": operator.mul,
    f"{element[3]}": operator.truediv
    }   
    print(ops)
    
    def calc(n):
        """
        Function to calculate our sum for varying values of n. In our construction n is a square number as we are
        Considering a matrix. 
        Consdier first:
        rightarrow=+
        leftarrow=-
        uparrow=x
        downarrow=divides
        """
        total=1
        for i in range(1, n + 1):
            if n==1:
                return 1
            if i % 2 == 1:
                for j in range(i ** 2 - i + 1, i ** 2 + 1):
                    #rightarrow
                    op_func = ops['rightarrow']
                    total = op_func(total, (j+1))
    #                 print(j)
    #                 print(f"total={total}")
    #                 #print(f"inside + and j={j}")

                for j in range(i ** 2 + 1,i ** 2 + i + 1):
                    # uparrow
                    if j > (n ** 2):
                        break
                    else:
                        op_func=ops['uparrow']
                        total = op_func(total, (j + 1))

    #                     print(j)
    #                     print(f"total={total}")

            elif i % 2 == 0:
                for j in range(i ** 2 - i + 1,i ** 2 + 1):
                    #leftarrow
                    op_func=ops['leftarrow']
                    total = op_func(total,j+1)

    #                 print(j)
    #                 print(f"total={total}")

                for j in range(i ** 2  + 1, i ** 2 + i + 1):
                    #downarrow
                    op_func=ops['downarrow']
                    total = op_func(total , (j + 1))
    #                 print(j)
    #                 print(f"total={total}")


        return total
    n_values=[2*i+1 for i in range(51)]
    res=[calc(n) for n in n_values]
    plt.plot(n_values,res)
    plt.title(f"rightarrow:{ops['rightarrow']},leftarrow:{ops['leftarrow']},uparrow:{ops['uparrow']},downarrow:{ops['downarrow']}")
    plt.show()
    print(calc(51))

        
```
